```python
from stackapi import StackAPI
```

```python
SITE = StackAPI('stackoverflow')
```

```python
SITE.page_size = 10
```

```python
SITE.max_pages = 10
```

```python
comments = SITE.fetch('comments', fromdate=1457136000, todate=1457222400,
    min=20, tagged='python', sort='votes')
comments
```

```python
questions = SITE.fetch('questions', fromdate=1457136000, todate=1457222400,
    min=20, tagged='python', sort='votes')
questions
```

```python
answers = SITE.fetch('answers', fromdate=1457136000, todate=1457222400,
    min=20, tagged='python', sort='votes')
answers
```
